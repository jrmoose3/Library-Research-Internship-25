```python
def sum(number_range1):
  previous_number = 0
  for i in range(number_range1):
    sum = previous_number + i
    print(f"Current Number: {i} Previous Number: {previous_number} Sum: {sum}")
    previous_number = i

sum(10)
```

```python
# write a code to return True if the list's first and last numbers are
# the same. If the numbers are different, return False

def true_or_false (list_name):
  for i in list_name:
    if list_name[0] == list_name[-1]:
      return True
    else:
      return False

numbers_x = [10, 20, 30, 40, 10]
numbers_y = [75, 65, 35, 75, 30]

true_or_false(numbers_y)
```

```python
def palindrome(number):
  original_number = str(number)
  reverse_number = original_number[::-1]
  if original_number == reverse_number:
    print("palindrome")
  else:
    print("no")


palindrome(121)
```
