```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the Excel file
df = pd.read_excel('Cranevelt_Final_Sheet.xlsx', sheet_name='1520-1528-Individual Months')

# Plotting the scatterplot and regression line
sns.regplot(x='Letters Sent', y='Unique Author Count', data=df, scatter_kws={'s': 50}, line_kws={'color': 'red'})

# Title and labels
plt.title('Letters Received against Monthly Authors')
plt.xlabel('Letters Received')
plt.ylabel('Total Monthly Authors')

# Calculate the correlation coefficient
correlation = df['Letters Sent'].corr(df['Unique Author Count'])

# Display the correlation coefficient
plt.text(0.05, 0.95, f'Correlation: {correlation:.2f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')

# Show the plot
plt.show()

# Print the correlation coefficient
print(f"Correlation coefficient: {correlation}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import mean_absolute_error

# Load the Excel data
df = pd.read_excel('Cranevelt_Final_Sheet.xlsx', sheet_name='1520-1528-Individual Months')

# Scatterplot + regression line
sns.regplot(x='Letters Sent', y='Unique Author Count', data=df,
            scatter_kws={'s': 50}, line_kws={'color': 'red'})

# Add y = x line (perfect match)
max_val = max(df['Letters Sent'].max(), df['Unique Author Count'].max())
plt.plot([0, max_val], [0, max_val], color='green', linestyle='--', label='Perfect Match (y = x)')

# Title and labels
plt.title('Letters Sent vs Unique Authors')
plt.xlabel('Letters Received')
plt.ylabel('Unique Author Count')

# Calculate correlation
correlation = df['Letters Sent'].corr(df['Unique Author Count'])

# Calculate exact match proportion
exact_matches = (df['Letters Sent'] == df['Unique Author Count']).sum()
total_points = len(df)
exact_match_pct = (exact_matches / total_points) * 100

# Calculate MAE
mae = mean_absolute_error(df['Letters Sent'], df['Unique Author Count'])

# Add text to plot

plt.text(0.05, 0.90, f'Exact Matches: {exact_matches} months out of 108 ({exact_match_pct:.1f}%)',
         transform=plt.gca().transAxes, fontsize=12)
plt.text(0.05, 0.85, f'Mean Absolute Error (MAE): {mae:.2f} letters', transform=plt.gca().transAxes, fontsize=12)

plt.legend()
plt.show()

# Also print in console
print(f"Correlation coefficient: {correlation:.2f}")
print(f"Exact Matches: {exact_matches} out of {total_points} ({exact_match_pct:.1f}%)")
print(f"Mean Absolute Error (MAE): {mae:.2f}")

```
