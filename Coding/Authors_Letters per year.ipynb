```python
# Load the data using pandas
import pandas as pd

# Replace with the excel data
df = pd.read_excel('Author_Counts.xlsx')

df.head()unique_authors_per_month_year = df.groupby(['Year', 'Date_Month'])['Author_Full_Name'].nunique()

# Convert to a dictionary
author_count_dict = unique_authors_per_month_year.to_dict()

# Reset index
author_counts_df = unique_authors_per_month_year.reset_index()

# Rename columns
author_counts_df.columns = ['Year', 'Month', 'Unique_Author_Count']

# create pivot format
pivot_table = author_counts_df.pivot(index='Year', columns='Month', values='Unique_Author_Count')

# Export to Excel
with pd.ExcelWriter('unique_authors_by_month_year.xlsx') as writer:
    author_counts_df.to_excel(writer, sheet_name='Flat_Table', index=False)
    pivot_table.to_excel(writer, sheet_name='Pivot_Table')

```

```python

# Load the data using pandas
import pandas as pd

# Replace with the excel data
df = pd.read_excel('Author_Counts.xlsx')

letters_per_author_year = df.groupby(['Year', 'Author_Full_Name']).size().reset_index(name='Letter Count')

# Convert to dictionary
from collections import defaultdict

author_year_dict = defaultdict(dict)
for _, row in letters_per_author_year.iterrows():
    year = row['Year']
    author = row['Author_Full_Name']
    count = row['Letter Count']
    author_year_dict[year][author] = count

# Print results
print("Letters sent per author by year:")
for year, authors in author_year_dict.items():
    print(f"\nYear: {year}")
    for author, count in authors.items():
        print(f"  {author}: {count}")
```

```python

# Create a pivot table
pivot_df = df.pivot_table(index='Author_Full_Name', columns='Year', aggfunc='size', fill_value=0)

# Reset index
pivot_df = pivot_df.reset_index().rename(columns={'Author': 'Name'})

# Show the result
print(pivot_df)
```

```python
# Pivot table to excel
pivot_df.to_excel("cranevelt_letters.xlsx", index=False)
```
